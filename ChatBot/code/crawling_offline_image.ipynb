### exel

```python
import pandas as pd

# 데이터를 'courseActiveSeq'와 'img URL' 컬럼으로 변환
records = []
for key, urls in url_image_merged.items():
    # 만약 URLs가 문자열로 되어 있다면 리스트로 변환
    if isinstance(urls, str):
        urls = [urls]  # 문자열을 리스트로 변환
    
    for url in urls:
        records.append([key, url])

# 데이터프레임 생성
df = pd.DataFrame(records, columns=['courseActiveSeq', 'img URL'])

# 엑셀 파일로 저장
df.to_excel('data_edu_img_url_merged.xlsx', index=False)

# 데이터프레임 확인 (optional)
print(df)
```
