# 정보 크롤링 시작

```python
import json

with open(r'C:\sesac\project3\offline_courses_url.json', "r", encoding="utf-8") as file:
    url_ls = json.load(file)
print(url_ls.keys())
print(len(url_ls))
```

```python
import requests
from bs4 import BeautifulSoup
import re
import time


offline_courses_info = {}

for i in url_ls.keys():
    if i == '1650':
        time.sleep(5)
    
    print(i)
    response = requests.get(f"https://sesac.seoul.kr/course/active/detail.do?courseActiveSeq={i}")
    soup = BeautifulSoup(response.text, 'html.parser')

    # 강의 제목
    title_element = soup.select_one('div.edu_on_detail p.sub')
    title = title_element.text.strip() if title_element else "정보 없음"

    #강의 분류
    content_classify = soup.select('span.cs_type')[1].text.strip() if len(soup.select('span.cs_type')) > 1 else "정보 없음"

    # 지원기간, 수업기간, 수업시간
    learning_info = soup.select('div.adm_info span.val')
    learning_ls = [info.text.strip() for info in learning_info]

    application_period = learning_ls[0] if len(learning_ls) > 0 else "정보 없음"
    training_period = learning_ls[1] if len(learning_ls) > 1 else "정보 없음"
    training_hour = learning_ls[2] if len(learning_ls) > 2 else "정보 없음"

    # 교육장소, 수강대상, 신청방법, 선발기준, 교육목표, 이수기준
    group_ls = [group.text.strip() for group in soup.select('#c_01 > div.group')]
    
    structured_group_data = {}
    for item in group_ls:
        parts = re.split(r"\n+", item, maxsplit=1)  # 제목과 내용 분리
        if len(parts) == 2:
            key, value = parts
            structured_group_data[key.strip()] = value.strip().replace('\n', '').replace('\t', '').replace('\xa0','')

    # 이수기준 키 존재 여부 확인 후 처리
    if '이수기준' in structured_group_data:
        structured_group_data['이수기준'] = structured_group_data['이수기준'].replace(' ', '')

    # 강의목차
    content_table_data = [
        k.text.strip().replace('\xa0', '').replace('\n', '').replace('\t', '')
        for k in soup.select('table.tbl_item tr')
    ]

    # 최종 결과 저장
    offline_courses_info[i] = {
        "강의제목": title,
        "강의분류": content_classify,
        "지원기간": application_period,
        "수업기간": training_period,
        "수업시간": training_hour,
        "세부정보": structured_group_data,
        "강의목차": content_table_data
    }

    # break  # 첫 번째 강의만 가져오도록 유지

    print(f'{i}url 완료')
    print(offline_courses_info[i])
```

## 크롤링결과 저장

```python
json_filename = "offline_courses.json"
with open(json_filename, "w", encoding="utf-8") as json_file:
    json.dump(offline_courses_info, json_file, ensure_ascii=False, indent=4)

print(f"데이터가 '{json_filename}' 파일로 저장되었습니다.")
```
